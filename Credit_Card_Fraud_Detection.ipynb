## **Model building on imbalanced data**

**Metric selection for heavily imbalanced data. Here are some Noteworthy points to be discussed.**

- The dataset is **heavily imbalanced**, with only about **0.17% of transactions being fraudulent**.  
  Therefore, **accuracy** is not a reliable metric for evaluating the model.  

- For instance, if a model predicts the same class (all `0`s or all `1`s) for every data point, it would still achieve an accuracy of over **99%**, while completely failing to detect fraud.  

- To fairly evaluate the model, we use the **ROC-AUC score**, which considers performance across all classification thresholds.  

- In other words, **accuracy alone is misleading** in imbalanced datasets, as it does not capture the model's ability to correctly identify the **minority class**.  

- The **ROC curve** helps us assess the modelâ€™s performance across thresholds. It shows the trade-off between **True Positive Rate (TPR)** and **False Positive Rate (FPR)**.  

- The default decision threshold of **0.5** is not always optimal.  
  - The best threshold is typically the one where **TPR is high** and **FPR is low**, minimizing misclassifications.  

- Once the **optimal threshold** is chosen, we can compute the **F1 score**, which balances **precision** and **recall**, to better measure the classifierâ€™s effectiveness at that threshold.  


**Why SVM (and in some cases Random Forest) was not used for model building?**

- The dataset contains **284,807 data points**, and after applying **oversampling**, the size becomes even larger.  
- **Support Vector Machines (SVMs)** are not efficient on very large datasets since they require substantial **computational power** and **memory** to perform the transformations.  
- Performing **K-Fold Cross Validation** for hyperparameter tuning further increases the computational cost and becomes extremely **time-consuming**.  
- Due to these **resource and time constraints**, SVM was not considered for this problem.  

- Similarly, in a few cases, **Random Forest** was also avoided because training multiple deep decision trees on such a large dataset with cross-validation can be computationally expensive and slow.  


**Why KNN was not used for model building?**

- **K-Nearest Neighbors (KNN)** is not memory efficient, as the algorithm needs to **store all data points** during training.  
- As the dataset size increases, KNN becomes **very slow**, since for every new data point it must compute the **distance to all other points** in the dataset to find the nearest neighbors.  
- This makes KNN **computationally expensive** and impractical for large datasets like ours.

**Tuning Hyperparameter (C in Regularization)**  

- **Regularization** is used to prevent overfitting by adding a **penalty term** to the loss function, discouraging the model from becoming overly complex and fitting the training data too closely.  

- The hyperparameter **C** controls the strength of regularization:  
  - **Smaller C** â†’ Stronger regularization  
    - The model is more constrained, leading to **simpler decision boundaries**.  
    - Helps reduce overfitting, but if too small, it can cause **underfitting**.  
  - **Larger C** â†’ Weaker regularization  
    - The model fits the training data more closely.  
    - May improve training performance, but increases the risk of **overfitting**.  


```python
# Importing libraries for cross validation
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
# Creating KFold object with 5 splits
folds = KFold(n_splits=5, random_state=4, shuffle=True)

# specify params
params = {'C': [0.01, 0.1, 1, 10, 100, 1000]}

# Specifing score as recall as we are more focused on acheiving the higher sensitivity than the accuracy
model_cv = GridSearchCV(estimator = LogisticRegression(), 
                        param_grid = params, 
                        scoring ='roc_auc', 
                        cv=folds, 
                        verbose=1, 
                        return_train_score=True)
```

```python
# fit the model
model_cv.fit(X_train, y_train)
```

```python
# results of GridSearchCV
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
cv_results[['params', 'mean_test_score', 'std_test_score', 'rank_test_score', 'mean_fit_time', 'mean_score_time', 'mean_train_score']]
```

```python
# plot of C versus train and validation scores

plt.figure(figsize=(8, 6))
plt.plot(cv_results['param_C'], cv_results['mean_test_score'])
plt.plot(cv_results['param_C'], cv_results['mean_train_score'])
plt.xlabel('C')
plt.ylabel('roc_auc')
plt.legend(['test result', 'train result'], loc='upper right')
plt.xscale('log')
```

```python
# best score
best_score = model_cv.best_score_
best_score
```

```python
# best params
best_params = model_cv.best_params_['C']
best_params
```

**ROC on the train set**

```python
# ROC Curve function

def draw_roc( actual, probs ):
    fpr, tpr, thresholds = metrics.roc_curve( actual, probs,
                                              drop_intermediate = False )
    auc_score = metrics.roc_auc_score( actual, probs )
    plt.figure(figsize=(5, 5))
    plt.plot( fpr, tpr, label='ROC curve (area = %0.2f)' % auc_score )
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
    plt.ylabel('True Positive Rate')
    plt.title('Receiver operating characteristic example')
    plt.legend(loc="lower right")
    plt.show()

    return None
```

**Hyperparameter Tuning**

```python
# hyperparameter tuning with XGBoost

# creating a KFold object 
folds = 3

# specify range of hyperparameters
param_grid = {'learning_rate': [0.2, 0.6],
               'subsample': [0.3, 0.6, 0.9]}

# specify model
xgb_model = XGBClassifier(max_depth=2, n_estimators=200)

# set up GridSearchCV()
model_cv = GridSearchCV(estimator = xgb_model, 
                        param_grid = param_grid, 
                        scoring= 'roc_auc', 
                        cv = folds, 
                        verbose = 1,
                        return_train_score=True)

# fit the model
model_cv.fit(X_train, y_train)
```

```python
cv_results = pd.DataFrame(model_cv.cv_results_)
cv_results
```

```python
# # plotting
plt.figure(figsize=(16,6))

param_grid = {'learning_rate': [0.2, 0.6], 
             'subsample': [0.3, 0.6, 0.9]} 


for n, subsample in enumerate(param_grid['subsample']):
    

    # subplot 1/n
    plt.subplot(1,len(param_grid['subsample']), n+1)
    df = cv_results[cv_results['param_subsample']==subsample]

    plt.plot(df["param_learning_rate"], df["mean_test_score"])
    plt.plot(df["param_learning_rate"], df["mean_train_score"])
    plt.xlabel('learning_rate')
    plt.ylabel('AUC')
    plt.title("subsample={0}".format(subsample))
    plt.ylim([0.60, 1])
    plt.legend(['test score', 'train score'], loc='best')
    plt.xscale('log')
```

# **Handling Class Imbalance**

As we can see, the data is heavily imbalanced.  
We will try several approaches to address this issue.


**Resampling Techniques**

- **Undersampling**  
  Randomly reduce the majority class instances to match the minority class size.  

- **Oversampling**  
  Duplicate or generate synthetic instances of the minority class to match the majority class size.  

- **SMOTE (Synthetic Minority Over-sampling Technique)**  
  Create synthetic minority class instances by interpolating between neighboring instances.  

- **ADASYN (Adaptive Synthetic Sampling)**  
  Generate synthetic instances in regions where the minority class is difficult to learn.  


**Why Undersampling is not Preferred in Industry**

- **Loss of information**: Undersampling reduces the size of the majority class, potentially discarding important patterns and leading to decreased performance on that class.  

- **Incomplete understanding of the problem**: By removing data points from the majority class, we risk losing valuable insights that may contribute to the overall predictive power of the model.  

- **Risk with small datasets**: If the dataset is already small, undersampling further reduces the dataset size, which can increase the chance of **overfitting** and result in poor generalization performance.  


**SMOTE** **(Synthetic Minority Oversampling Technique)**

## **Cost-Benefit Analysis**

- We have tried several models with both **balanced** and **imbalanced** data.  
  Most of the models have performed reasonably well in terms of **ROC score, Precision, and Recall**.  

- While selecting the **best model**, we must consider factors such as:  
  - Availability of **infrastructure** and **resources**  
  - Required **computational power** to train and deploy the model  

- **Complex models** such as **Random Forest, SVM, and XGBoost** require heavy computational resources.  
  Building and maintaining such infrastructure increases the **deployment cost**.  

- **Simpler models** such as **Logistic Regression** require fewer resources, making them more cost-effective.  

- A critical factor is the **trade-off between model performance and cost**.  
  - If a small improvement in **ROC score** leads to **significant monetary gain or loss** (e.g., in banking or fraud detection), it may justify using a **complex model**, despite its higher computational cost.  


# **Summary to the Business**

- For banks with **smaller average transaction value**, we should focus on achieving **high Precision**.  
  - Rationale: We only want to label relevant transactions as fraudulent.  
  - Each flagged transaction can be verified by involving the **human element** (e.g., calling the customer).  
  - However, if precision is **low**, this leads to many **false positives**, which increase the operational burden on humans.  

- For banks with **larger average transaction value**, the focus shifts to **high Recall**.  
  - Rationale: If recall is **low**, the model fails to detect many actual fraudulent transactions (**false negatives**).  
  - In such cases, the potential **loss from missing a high-value fraudulent transaction** can be very significant.  

- Therefore, for **high-value transactions**, the bank should prioritize **high Recall** in order to minimize undetected fraud.  

- After evaluating several models, we observed that with the **SMOTE-balanced dataset**,  
  **Logistic Regression** achieved:  
  - A strong **ROC score**  
  - High **Recall**  
  - Easy **interpretability** and **explainability** for business stakeholders
